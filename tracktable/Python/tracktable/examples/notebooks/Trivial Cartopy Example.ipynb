```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
#ax.coastlines()
ax.stock_img()

ny_lon, ny_lat = -75, 43
delhi_lon, delhi_lat = 77.23, 28.61

plt.plot([delhi_lon, ny_lon], [delhi_lat, ny_lat],
        color='blue',
        linewidth=2,
        marker='o',
        transform=ccrs.Geodetic())
```
